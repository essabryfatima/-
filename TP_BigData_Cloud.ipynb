```python
import pandas as pd

# Télécharger le dataset Titanic directement depuis Internet
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
data = pd.read_csv(url)

# Afficher les 5 premières lignes
data.head(5)
```

```python
# Description des données
data.describe()

# Colonnes et lignes
print("Nombre de lignes :", data.shape[0])
print("Nombre de colonnes :", data.shape[1])

# Exemple : âge moyen des passagers
print("Âge moyen :", data['Age'].mean())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.histplot(data['Age'].dropna(), kde=True)
plt.title("Distribution des âges dans le dataset Titanic")
plt.show()
```
